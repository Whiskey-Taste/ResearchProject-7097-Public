```python
# =============================================================================
# Complete Enhanced BERT Models with Dynamic Fusion and GPU Optimization
# Version: 5.0 - FIXED VERSION with proper class balancing and training
# =============================================================================

import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset, WeightedRandomSampler
import uuid
import time
import warnings
import re
import json
import pickle
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import (
    accuracy_score, f1_score, roc_auc_score, precision_score, 
    recall_score, confusion_matrix, classification_report
)
from sklearn.utils.class_weight import compute_class_weight
from transformers import (
    AutoTokenizer, AutoModelForSequenceClassification, 
    Trainer, TrainingArguments, EarlyStoppingCallback,
    get_linear_schedule_with_warmup, DataCollatorWithPadding
)
from datasets import Dataset
from typing import List, Dict, Optional, Tuple, Union, Any
from datetime import datetime
from abc import ABC, abstractmethod
from collections import defaultdict
import matplotlib.pyplot as plt
import seaborn as sns

# Haystack imports for RAG
from haystack import Document
from haystack.components.embedders import SentenceTransformersDocumentEmbedder, SentenceTransformersTextEmbedder
from haystack.document_stores.in_memory import InMemoryDocumentStore
from haystack.components.retrievers.in_memory import InMemoryEmbeddingRetriever
from haystack.document_stores.types import DuplicatePolicy

# LangChain imports
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.schema import BaseOutputParser

warnings.filterwarnings("ignore", category=DeprecationWarning)
os.environ["OPENAI_API_KEY"] = "Your-API-Key"


# =============================================================================
# Enhanced Device Manager with Memory Optimization
# =============================================================================

class EnhancedDeviceManager:
    """Enhanced CUDA device management with memory optimization"""
    
    @staticmethod
    def get_device():
        """Get the best available device with detailed info"""
        if torch.cuda.is_available():
            device = torch.device("cuda")
            gpu_name = torch.cuda.get_device_name(0)
            total_memory = torch.cuda.get_device_properties(0).total_memory / 1024**3  # GB
            
            print(f"üöÄ Using CUDA device: {gpu_name}")
            print(f"   Total Memory: {total_memory:.2f} GB")
            print(f"   Available Memory: {(total_memory - torch.cuda.memory_allocated(0) / 1024**3):.2f} GB")
            
            # Set memory optimization flags
            torch.cuda.empty_cache()
            torch.backends.cudnn.benchmark = True
            torch.backends.cudnn.deterministic = False
            
            return device
        else:
            print("üíª CUDA not available, using CPU")
            return torch.device("cpu")
    
    @staticmethod
    def optimize_batch_size(model_size_mb: float, available_memory_gb: float = 12.0) -> int:
        """Calculate optimal batch size based on model size and available memory"""
        usable_memory_gb = available_memory_gb - 4.0
        
        if model_size_mb < 500:
            base_batch_size = 32  # Reduced from 64
        elif model_size_mb < 1000:
            base_batch_size = 16  # Reduced from 32
        else:
            base_batch_size = 8   # Reduced from 16
        
        memory_factor = usable_memory_gb / 8.0
        optimal_batch_size = int(base_batch_size * memory_factor)
        
        return max(8, min(64, optimal_batch_size))  # Reduced max from 128
    
    @staticmethod
    def clear_cuda_cache():
        """Clear CUDA cache to free memory"""
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
            print("üßπ Cleared CUDA cache")

# Get device once at module level
DEVICE = EnhancedDeviceManager.get_device()

# =============================================================================
# Enhanced Base Classes
# =============================================================================

class ModelEvaluator:
    """Unified model evaluator with enhanced metrics"""
    
    @staticmethod
    def evaluate_predictions(y_true, y_pred, y_proba=None):
        """Calculate standard evaluation metrics"""
        metrics = {
            'accuracy': accuracy_score(y_true, y_pred),
            'f1': f1_score(y_true, y_pred, zero_division=0),
            'precision': precision_score(y_true, y_pred, zero_division=0),
            'recall': recall_score(y_true, y_pred, zero_division=0)
        }
        
        if y_proba is not None:
            try:
                metrics['roc_auc'] = roc_auc_score(y_true, y_proba)
            except:
                metrics['roc_auc'] = 0.5
        
        return metrics

class EnhancedDataProcessor:
    """Enhanced data processor for handling larger datasets efficiently"""
    
    @staticmethod
    def load_and_prepare_data(file_path="./final_processed_data.csv", 
                            sample_size: Optional[int] = None,
                            balance_classes: bool = True,
                            min_samples_per_class: int = 2000) -> pd.DataFrame:
        """Load and prepare FAERS data with optional sampling and class balancing"""
        print(f"üìÇ Loading data from {file_path}...")
        df = pd.read_csv(file_path)
        
        # Filter valid dechallenge values
        df = df[df["dechal"].isin(["Y", "N"])].copy()
        
        # Create label
        df["label"] = df["dechal"].map({"Y": 1, "N": 0})
        
        # Create comprehensive text representation
        df["text"] = (
            "Symptoms: " + df["pt"].fillna("Unknown") +
            " | Indications: " + df["indi_pt"].fillna("None") +
            " | Drug: " + df["drugname_clean"].fillna("Unknown") +
            " | Age: " + df["age"].fillna("Unknown").astype(str) +
            " | Sex: " + df["sex"].fillna("Unknown")
        )
        
        # Balance classes if needed
        if balance_classes:
            # Find minority class size
            label_counts = df["label"].value_counts()
            minority_size = label_counts.min()
            
            # Check if we have enough samples
            if minority_size < min_samples_per_class:
                print(f"‚ö†Ô∏è  Warning: Minority class has only {minority_size} samples")
                print(f"   Recommended minimum: {min_samples_per_class} samples per class")
            
            # Undersample majority class - allow up to 2:1 ratio
            balanced_dfs = []
            for label in [0, 1]:
                label_df = df[df["label"] == label]
                if len(label_df) > minority_size * 2:  
                    label_df = label_df.sample(n=minority_size * 2, random_state=42)
                balanced_dfs.append(label_df)
            
            df = pd.concat(balanced_dfs, ignore_index=True)
            df = df.sample(frac=1, random_state=42).reset_index(drop=True)  # Shuffle
        
        # Sample if needed
        if sample_size and len(df) > sample_size:
            df, _ = train_test_split(
                df,
                train_size=sample_size,
                stratify=df["label"],
                random_state=42
            )
        
        print(f"‚úÖ Loaded {len(df)} records")
        print(f"   Label distribution: {df['label'].value_counts().to_dict()}")
        print(f"   Class ratio: {df['label'].mean():.2%} positive")
        
        return df
    
    @staticmethod
    def create_enhanced_splits(df: pd.DataFrame, 
                             test_size: float = 0.15,
                             val_size: float = 0.15,
                             random_state: int = 42) -> Tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame]:
        """Create train/val/test splits with stratification"""
        train_val_df, test_df = train_test_split(
            df[["text", "label"]], 
            test_size=test_size, 
            stratify=df["label"], 
            random_state=random_state
        )
        
        val_size_adjusted = val_size / (1 - test_size)
        train_df, val_df = train_test_split(
            train_val_df,
            test_size=val_size_adjusted,
            stratify=train_val_df["label"],
            random_state=random_state
        )
        
        print(f"üìä Data splits:")
        print(f"   Train: {len(train_df)} samples (pos: {train_df['label'].mean():.2%})")
        print(f"   Val: {len(val_df)} samples (pos: {val_df['label'].mean():.2%})") 
        print(f"   Test: {len(test_df)} samples (pos: {test_df['label'].mean():.2%})")
        
        return train_df, val_df, test_df

# =============================================================================
# Custom Trainer with Class Weights
# =============================================================================

class WeightedTrainer(Trainer):
    """Custom trainer that supports class weights"""
    
    def __init__(self, class_weights=None, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.class_weights = class_weights
        
    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None):
        """Compute weighted loss - Fixed to handle new parameter"""
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits
        
        if self.class_weights is not None:
            # Convert class weights to tensor
            weights = torch.tensor(self.class_weights, device=model.device, dtype=torch.float32)
            loss_fct = nn.CrossEntropyLoss(weight=weights)
        else:
            loss_fct = nn.CrossEntropyLoss()
            
        loss = loss_fct(logits.view(-1, self.model.config.num_labels), labels.view(-1))
        
        return (loss, outputs) if return_outputs else loss

# =============================================================================
# Dynamic Fusion Network
# =============================================================================

class DynamicFusionNetwork(nn.Module):
    """Neural network for learning optimal component weights dynamically"""
    
    def __init__(self, num_components: int = 4, hidden_dim: int = 128):
        super().__init__()
        self.num_components = num_components
        
        input_dim = num_components * 3
        
        self.fusion_network = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim // 2, num_components),
            nn.Softmax(dim=-1)
        )
        
        self.confidence_gate = nn.Sequential(
            nn.Linear(num_components, hidden_dim // 2),
            nn.ReLU(),
            nn.Linear(hidden_dim // 2, num_components),
            nn.Sigmoid()
        )
    
    def forward(self, component_features: torch.Tensor) -> Tuple[torch.Tensor, torch.Tensor]:
        weights = self.fusion_network(component_features)
        confidences = component_features[:, 2::3]
        gates = self.confidence_gate(confidences)
        gated_weights = weights * gates
        gated_weights = gated_weights / (gated_weights.sum(dim=1, keepdim=True) + 1e-8)
        return gated_weights, gates

# =============================================================================
# Model 1: Enhanced BERT Baseline
# =============================================================================

class EnhancedBERTBaseline:
    """Enhanced BERT baseline with optimizations and class weight support"""
    
    def __init__(self, model_name='bert-base-uncased', num_labels=2):
        self.device = DEVICE
        self.model_name = model_name
        self.num_labels = num_labels
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = AutoModelForSequenceClassification.from_pretrained(
            model_name, num_labels=num_labels
        ).to(self.device)
        self.trainer = None
        self.class_weights = None
        
        model_size_mb = sum(p.numel() * p.element_size() for p in self.model.parameters()) / 1024**2
        self.optimal_batch_size = EnhancedDeviceManager.optimize_batch_size(model_size_mb)
        print(f"üì¶ Optimal batch size: {self.optimal_batch_size}")
    
    def tokenize_data(self, df: pd.DataFrame, max_length: int = 256):
        """Tokenize data with fixed num_proc=1 to avoid pickle issues"""
        dataset = Dataset.from_pandas(df)
        
        def tokenize_function(examples):
            return self.tokenizer(
                examples['text'], 
                truncation=True, 
                padding='max_length', 
                max_length=max_length
            )
        
        return dataset.map(
            tokenize_function, 
            batched=True,
            num_proc=1  # Fixed: avoid multiprocessing issues
        )
    
    def train(self, train_df: pd.DataFrame, eval_df: Optional[pd.DataFrame] = None, 
              epochs: int = 5, learning_rate: float = 3e-5,  # Reduced learning rate
              warmup_steps: int = 500, patience: int = 3):
        """Train with early stopping, class weights, and optimizations"""
        
        # Calculate class weights
        labels = train_df['label'].values
        class_weights = compute_class_weight('balanced', classes=np.unique(labels), y=labels)
        self.class_weights = class_weights
        print(f"üìä Class weights: {dict(zip([0, 1], class_weights))}")
        
        train_dataset = self.tokenize_data(train_df)
        eval_dataset = self.tokenize_data(eval_df) if eval_df is not None else None
        
        # Data collator for dynamic padding
        data_collator = DataCollatorWithPadding(self.tokenizer, padding=True)
        
        # Adjust batch size based on data size
        train_batch_size = min(self.optimal_batch_size, 16) if len(train_df) < 5000 else self.optimal_batch_size
        
        training_args = TrainingArguments(
            output_dir=f'./enhanced_{self.model_name}_results',
            num_train_epochs=epochs,
            per_device_train_batch_size=train_batch_size,
            per_device_eval_batch_size=train_batch_size * 2,
            learning_rate=learning_rate,
            warmup_steps=min(warmup_steps, len(train_dataset) // train_batch_size // 2),
            weight_decay=0.01,
            logging_steps=50,
            eval_strategy="steps" if eval_dataset else "no",
            eval_steps=100 if eval_dataset else None,
            save_strategy="steps",
            save_steps=200,
            load_best_model_at_end=True if eval_dataset else False,
            metric_for_best_model="f1" if eval_dataset else None,
            greater_is_better=True if eval_dataset else None,
            fp16=torch.cuda.is_available(),
            gradient_checkpointing=True,
            gradient_accumulation_steps=2,  # Added for stability
            dataloader_num_workers=0,
            remove_unused_columns=True,
            label_names=["labels"],
            eval_accumulation_steps=1,  # Added to prevent OOM during eval
            push_to_hub=False,
            report_to="none",  # Disable wandb/tensorboard reporting
        )
        
        def compute_metrics(eval_pred):
            logits, labels = eval_pred
            predictions = np.argmax(logits, axis=-1)
            probs = torch.softmax(torch.tensor(logits), dim=-1).numpy()
            metrics = ModelEvaluator.evaluate_predictions(labels, predictions, probs[:, 1])
            
            # Add detailed metrics
            cm = confusion_matrix(labels, predictions)
            if cm.shape == (2, 2):
                metrics['true_negatives'] = cm[0, 0]
                metrics['false_positives'] = cm[0, 1] 
                metrics['false_negatives'] = cm[1, 0]
                metrics['true_positives'] = cm[1, 1]
            
            return metrics
        
        callbacks = []
        if eval_dataset and patience > 0:
            callbacks.append(EarlyStoppingCallback(
                early_stopping_patience=patience,
                early_stopping_threshold=0.001
            ))
        
        # Use custom trainer with class weights
        self.trainer = WeightedTrainer(
            class_weights=self.class_weights,
            model=self.model,
            args=training_args,
            train_dataset=train_dataset,
            eval_dataset=eval_dataset,
            compute_metrics=compute_metrics if eval_dataset else None,
            callbacks=callbacks,
            data_collator=data_collator,
        )
        
        train_result = self.trainer.train()
        self.trainer.save_model(f"./best_{self.model_name}_model")
        return train_result
    
    def predict_with_confidence(self, texts: List[str]) -> Dict[str, List]:
        """Fixed: Returns predictions and probabilities with consistent format"""
        if isinstance(texts, str):
            texts = [texts]
            
        self.model.eval()
    
        # Process in batches to avoid OOM
        batch_size = min(32, self.optimal_batch_size)
        all_preds = []
        all_probs = []
        all_confidences = []
        
        for i in range(0, len(texts), batch_size):
            batch_texts = texts[i:i+batch_size]
            
            inputs = self.tokenizer(
                batch_texts,
                return_tensors="pt",
                padding=True,
                truncation=True,
                max_length=512
            )
            inputs = {k: v.to(self.device) for k, v in inputs.items()}
        
            with torch.no_grad():
                outputs = self.model(**inputs)
                logits = outputs.logits
                probs = torch.softmax(logits, dim=1)
                preds = torch.argmax(logits, dim=1)
            
            all_preds.extend(preds.detach().cpu().numpy().tolist())
            all_probs.extend(probs[:, 1].detach().cpu().numpy().tolist())
            all_confidences.extend(torch.max(probs, dim=1)[0].detach().cpu().numpy().tolist())
    
        # Fixed: Ensure consistent output format
        return {
            "preds": all_preds,
            "probs": all_probs,  # Class 1 probabilities
            "predictions": all_preds,
            "probabilities": all_probs,
            "confidences": all_confidences
        }

    def evaluate(self, test_df: pd.DataFrame) -> Dict:
        """Evaluate model performance"""
        if self.trainer is None:
            raise ValueError("Model must be trained before evaluation")
        
        test_dataset = self.tokenize_data(test_df)
        results = self.trainer.evaluate(test_dataset)
        return results

# =============================================================================
# Model 2: Biomedical BERT Variants
# =============================================================================

class BiomedicalBERT(EnhancedBERTBaseline):
    """Biomedical BERT models with all variants"""
    
    BIOMEDICAL_MODELS = {
        'BioBERT': 'dmis-lab/biobert-base-cased-v1.2',
        'ClinicalBERT': 'emilyalsentzer/Bio_ClinicalBERT',
        'SciBERT': 'allenai/scibert_scivocab_uncased',
        'PubMedBERT': 'microsoft/BiomedBERT-base-uncased-abstract-fulltext',
        'BlueBERT': 'bionlp/bluebert_pubmed_mimic_uncased_L-12_H-768_A-12',
        'BioRoBERTa': 'allenai/biomed_roberta_base'
    }
    
    def __init__(self, model_type='BioBERT', num_labels=2):
        if model_type not in self.BIOMEDICAL_MODELS:
            raise ValueError(f"Model type must be one of {list(self.BIOMEDICAL_MODELS.keys())}")
        
        model_name = self.BIOMEDICAL_MODELS[model_type]
        try:
            super().__init__(model_name, num_labels)
            self.model_type = model_type
            print(f"üß¨ Initialized {model_type} from {model_name}")
        except Exception as e:
            print(f"‚ö†Ô∏è Failed to load {model_type}, falling back to BioBERT")
            super().__init__(self.BIOMEDICAL_MODELS['BioBERT'], num_labels)
            self.model_type = 'BioBERT'

# =============================================================================
# Enhanced RAG Module
# =============================================================================

class EnhancedRAGModule:
    """Enhanced RAG with confidence-based retrieval and batch processing"""
    
    def __init__(self, embedding_model="sentence-transformers/all-MiniLM-L6-v2"):
        self.embedding_model = embedding_model
        
        self.doc_embedder = SentenceTransformersDocumentEmbedder(model=embedding_model)
        self.text_embedder = SentenceTransformersTextEmbedder(model=embedding_model)
        
        self.document_store = InMemoryDocumentStore()
        self.retriever = InMemoryEmbeddingRetriever(document_store=self.document_store)
        
        self.is_initialized = False
        self.retrieval_stats = defaultdict(list)
    
    def build_knowledge_base(self, df: pd.DataFrame, batch_size: int = 500):
        """Build knowledge base with batching for large datasets"""
        print(f"üîß Building enhanced RAG knowledge base...")
        
        total_docs = len(df)
        num_batches = (total_docs + batch_size - 1) // batch_size
        
        all_docs = []
        for i in range(num_batches):
            start_idx = i * batch_size
            end_idx = min((i + 1) * batch_size, total_docs)
            batch_df = df.iloc[start_idx:end_idx]
            
            batch_docs = [
                Document(
                    content=row['text'], 
                    meta={
                        'label': str(row['label']),
                        'confidence': 1.0,
                        'index': str(start_idx + idx)
                    }, 
                    id=str(uuid.uuid4())
                )
                for idx, (_, row) in enumerate(batch_df.iterrows())
            ]
            
            all_docs.extend(batch_docs)
            
            if (i + 1) % 10 == 0:
                print(f"   Processed {end_idx}/{total_docs} documents...")
        
        self.doc_embedder.warm_up()
        self.text_embedder.warm_up()
        
        print("   Generating embeddings...")
        embedded_docs = self.doc_embedder.run(all_docs)["documents"]
        
        self.document_store.write_documents(embedded_docs, policy=DuplicatePolicy.OVERWRITE)
        
        self.is_initialized = True
        print(f"‚úÖ Knowledge base built with {len(all_docs)} documents")
    
    def intelligent_retrieval(self, query: str, top_k: int = 5, 
                            confidence_threshold: float = 0.7) -> Dict:
        """Intelligent retrieval with confidence scoring"""
        if not self.is_initialized:
            raise ValueError("Knowledge base not initialized")
        
        query_embedding = self.text_embedder.run(query)["embedding"]
        
        results = self.retriever.run(
            query_embedding=query_embedding, 
            top_k=top_k * 2
        )
        
        similar_docs = results["documents"]
        
        filtered_docs = []
        for doc in similar_docs[:top_k]:
            doc_confidence = float(doc.meta.get('confidence', 1.0))
            if doc_confidence >= confidence_threshold:
                filtered_docs.append(doc)
        
        if len(filtered_docs) < 3:
            filtered_docs = similar_docs[:top_k]
        
        labels = [int(doc.meta['label']) for doc in filtered_docs]
        label_counts = {0: labels.count(0), 1: labels.count(1)}
        
        weights = [float(doc.meta.get('confidence', 1.0)) for doc in filtered_docs]
        weighted_score = sum(w * l for w, l in zip(weights, labels)) / sum(weights) if weights else 0.5
        
        self.retrieval_stats['query_length'].append(len(query.split()))
        self.retrieval_stats['docs_retrieved'].append(len(filtered_docs))
        self.retrieval_stats['recommendation_score'].append(weighted_score)
        
        return {
            'similar_cases': filtered_docs,
            'label_distribution': label_counts,
            'recommendation_score': weighted_score,
            'confidence_scores': weights,
            'retrieval_confidence': np.mean(weights) if weights else 0.5,
            'context_text': "\n".join([
                f"Case {i+1}: {doc.content} (Label: {doc.meta['label']})"
                for i, doc in enumerate(filtered_docs)
            ])
        }

# =============================================================================
# Model 3: BERT + RAG
# =============================================================================

class BERTWithRAG(EnhancedBERTBaseline):
    """BERT + RAG enhanced model with dynamic weighting"""
    
    def __init__(self, model_name='bert-base-uncased', num_labels=2, 
                 embedding_model="sentence-transformers/all-MiniLM-L6-v2"):
        super().__init__(model_name, num_labels)
        self.rag_module = EnhancedRAGModule(embedding_model)
        self.combination_strategy = "adaptive_weighted"
        self.rag_weight_low_conf = 0.4   # Higher RAG weight when BERT is uncertain
        self.rag_weight_high_conf = 0.2  # Lower RAG weight when BERT is confident
        
    def build_rag_knowledge_base(self, knowledge_df: pd.DataFrame):
        """Build RAG knowledge base"""
        self.rag_module.build_knowledge_base(knowledge_df)
    
    def predict_with_rag(self, texts: List[str]) -> Dict:
        """Predict with adaptive RAG enhancement"""
        if isinstance(texts, str):
            texts = [texts]
        
        # BERT predictions
        bert_results = self.predict_with_confidence(texts)
        
        # RAG retrieval
        rag_results = []
        for text in texts:
            retrieval_context = self.rag_module.intelligent_retrieval(text)
            rag_results.append(retrieval_context)
        
        # Adaptive combination based on BERT confidence
        enhanced_predictions = []
        enhanced_probabilities = []
        
        for i, text in enumerate(texts):
            bert_prob = bert_results['probs'][i]
            bert_confidence = bert_results['confidences'][i]
            rag_score = rag_results[i]['recommendation_score']
            
            # Adaptive weighting: use more RAG when BERT is uncertain
            if bert_confidence < 0.7:  # Low confidence threshold
                use_retrieval_weight = self.rag_weight_low_conf
            else:
                use_retrieval_weight = self.rag_weight_high_conf
            
            # Convert to probability arrays
            bert_prob_array = np.array([1 - bert_prob, bert_prob])
            rag_prob_array = np.array([1 - rag_score, rag_score])
            
            combined_prob = (1 - use_retrieval_weight) * bert_prob_array + use_retrieval_weight * rag_prob_array
            
            enhanced_probabilities.append(combined_prob)
            enhanced_predictions.append(np.argmax(combined_prob))
        
        return {
            'predictions': np.array(enhanced_predictions),
            'probabilities': np.array(enhanced_probabilities),
            'preds': enhanced_predictions,
            'probs': [prob[1] for prob in enhanced_probabilities],
            'bert_results': bert_results,
            'rag_results': rag_results,
            'adaptive_weights': [self.rag_weight_low_conf if conf < 0.7 else self.rag_weight_high_conf 
                               for conf in bert_results['confidences']]
        }

# =============================================================================
# LangChain Components
# =============================================================================

class EnhancedMedicalReasoningParser(BaseOutputParser):
    """Enhanced parser for medical reasoning outputs"""
    
    def parse(self, text: str) -> Dict:
        """Parse LLM reasoning output with enhanced extraction"""
        try:
            patterns = {
                'decision': r"(?:Decision|Recommendation):\s*([\w\s_]+?)(?:\n|$)",
                'confidence': r"(?:Confidence|Certainty):\s*(\d+(?:\.\d+)?)",
                'risk_level': r"(?:Risk_Level|Risk|Urgency):\s*(\w+)",
                'reasoning': r"(?:Reasoning|Explanation):\s*(.+?)(?=\n\n|\n[A-Z]|$)"
            }
            
            results = {}
            for key, pattern in patterns.items():
                match = re.search(pattern, text, re.IGNORECASE | re.DOTALL)
                if match:
                    results[key] = match.group(1).strip()
            
            decision_text = results.get('decision', '').lower()
            if 'discontinue' in decision_text or 'stop' in decision_text:
                decision = 1
            else:
                decision = 0
            
            confidence = float(results.get('confidence', '0.5'))
            if confidence > 1.0:
                confidence = confidence / 100.0
            
            risk_level = results.get('risk_level', 'Medium')
            
            return {
                'decision': decision,
                'confidence': confidence,
                'risk_level': risk_level,
                'reasoning': results.get('reasoning', text),
                'raw_output': text
            }
            
        except Exception as e:
            return {
                'decision': 0,
                'confidence': 0.5,
                'risk_level': 'Unknown',
                'reasoning': text,
                'raw_output': text,
                'parse_error': str(e)
            }

class EnhancedMedicalReasoningAgent:
    """Enhanced medical reasoning with dynamic chain selection"""
    
    def __init__(self, model_name: str = "gpt-3.5-turbo", temperature: float = 0.1):
        if not os.getenv("OPENAI_API_KEY"):
            raise ValueError("Please set OPENAI_API_KEY environment variable")
        
        self.llm = ChatOpenAI(model=model_name, temperature=temperature)
        self.parser = EnhancedMedicalReasoningParser()
        self.chain_usage_stats = defaultdict(int)
        self._setup_reasoning_chains()
    
    def _setup_reasoning_chains(self):
        """Setup multiple specialized reasoning chains"""
        self.chains = {}
        
        self.chains['standard'] = PromptTemplate(
            input_variables=['symptoms', 'indications', 'drug', 'bert_prediction', 
                           'bert_confidence', 'patient_factors'],
            template="""You are a medical AI assistant specializing in pharmacovigilance.

Patient Case:
- Symptoms: {symptoms}
- Indications: {indications}
- Drug: {drug}
- Patient Factors: {patient_factors}

Initial AI Assessment:
- BERT Prediction: {bert_prediction} (Confidence: {bert_confidence:.3f})

Based on medical knowledge and safety considerations, assess whether the drug should be discontinued.
Consider symptom severity, known adverse reactions, and risk-benefit ratio.

Provide a clinical assessment:

Decision: [Continue/Discontinue]
Confidence: [0.0-1.0]
Risk_Level: [Low/Medium/High]
Reasoning: [Detailed clinical reasoning]
"""
        )
    
    def adaptive_reasoning(self, case_data: Dict) -> Dict:
        """Perform adaptive reasoning based on case characteristics"""
        chain_name = 'standard'
        self.chain_usage_stats[chain_name] += 1
        
        inputs = {
            'symptoms': case_data['symptoms'],
            'indications': case_data['indications'],
            'drug': case_data['drug'],
            'bert_prediction': case_data['bert_prediction'],
            'bert_confidence': case_data['bert_confidence'],
            'patient_factors': case_data.get('patient_factors', 'Standard adult patient')
        }
        
        chain = self.chains[chain_name] | self.llm
        result = chain.invoke(inputs)
        
        parsed_result = self.parser.parse(result.content)
        parsed_result['reasoning_chain'] = chain_name
        
        return parsed_result

# =============================================================================
# Model 4: BERT + LangChain
# =============================================================================

class BERTWithLangChain(EnhancedBERTBaseline):
    """BERT + LangChain enhanced model with confidence-based combination"""
    
    def __init__(self, model_name: str = 'bert-base-uncased', num_labels: int = 2,
                 llm_model: str = "gpt-3.5-turbo", reasoning_strategy: str = "adaptive"):
        super().__init__(model_name, num_labels)
        self.reasoning_agent = EnhancedMedicalReasoningAgent(llm_model)
        self.reasoning_strategy = reasoning_strategy
        self.combination_method = "confidence_weighted"
    
    def parse_medical_text(self, text: str) -> Dict[str, str]:
        """Parse medical text fields"""
        patterns = {
            'symptoms': r'Symptoms:\s*([^|]+)',
            'indications': r'Indications:\s*([^|]+)',
            'drug': r'Drug:\s*([^|]+)',
            'age': r'Age:\s*([^|]+)',
            'sex': r'Sex:\s*([^|]+)'
        }
        
        results = {}
        for field, pattern in patterns.items():
            match = re.search(pattern, text)
            results[field] = match.group(1).strip() if match else ""
        
        return results
    
    def predict_with_reasoning(self, texts: List[str]) -> Dict:
        """Predict with confidence-aware reasoning enhancement"""
        if isinstance(texts, str):
            texts = [texts]
        
        # BERT predictions
        bert_results = self.predict_with_confidence(texts)
        
        # LangChain reasoning
        reasoning_results = []
        
        for i, text in enumerate(texts):
            medical_info = self.parse_medical_text(text)
            bert_pred = bert_results['preds'][i]
            bert_conf = bert_results['confidences'][i]
            
            case_data = {
                'symptoms': medical_info['symptoms'],
                'indications': medical_info['indications'],
                'drug': medical_info['drug'],
                'bert_prediction': "Discontinue" if bert_pred == 1 else "Continue",
                'bert_confidence': bert_conf,
                'patient_factors': f"Age: {medical_info['age']}, Sex: {medical_info['sex']}"
            }
            
            reasoning_result = self.reasoning_agent.adaptive_reasoning(case_data)
            reasoning_results.append(reasoning_result)
        
        # Confidence-weighted combination
        enhanced_predictions = []
        enhanced_probabilities = []
        
        for i in range(len(texts)):
            bert_prob = bert_results['probs'][i]
            bert_confidence = bert_results['confidences'][i]
            reasoning_decision = reasoning_results[i]['decision']
            reasoning_confidence = reasoning_results[i]['confidence']
            
            # Dynamic weighting based on confidence levels
            bert_weight = bert_confidence
            reasoning_weight = reasoning_confidence
            total_weight = bert_weight + reasoning_weight
            
            bert_weight_norm = bert_weight / total_weight
            reasoning_weight_norm = reasoning_weight / total_weight
            
            # Convert to probability arrays
            bert_prob_array = np.array([1 - bert_prob, bert_prob])
            
            if reasoning_decision == 1:
                reasoning_prob_array = np.array([1 - reasoning_confidence, reasoning_confidence])
            else:
                reasoning_prob_array = np.array([reasoning_confidence, 1 - reasoning_confidence])
            
            combined_prob = bert_weight_norm * bert_prob_array + reasoning_weight_norm * reasoning_prob_array
            enhanced_probabilities.append(combined_prob)
            enhanced_predictions.append(np.argmax(combined_prob))
        
        return {
            'predictions': np.array(enhanced_predictions),
            'probabilities': np.array(enhanced_probabilities),
            'preds': enhanced_predictions,
            'probs': [prob[1] for prob in enhanced_probabilities],
            'bert_results': bert_results,
            'reasoning_results': reasoning_results,
            'dynamic_weights': [(bert_results['confidences'][i], reasoning_results[i]['confidence']) 
                              for i in range(len(texts))]
        }

# =============================================================================
# Model 5: BERT + RAG + LangChain
# =============================================================================

class BERTWithRAGAndLangChain(EnhancedBERTBaseline):
    """BERT + RAG + LangChain model with triple fusion"""
    
    def __init__(self, model_name='bert-base-uncased', num_labels=2,
                 embedding_model="sentence-transformers/all-MiniLM-L6-v2",
                 llm_model="gpt-3.5-turbo"):
        super().__init__(model_name, num_labels)
        self.rag_module = EnhancedRAGModule(embedding_model)
        self.reasoning_agent = EnhancedMedicalReasoningAgent(llm_model)
        self.rag_initialized = False
        self.fusion_method = "confidence_based_fusion"
    
    def build_rag_knowledge_base(self, knowledge_df: pd.DataFrame):
        """Build RAG knowledge base"""
        self.rag_module.build_knowledge_base(knowledge_df)
        self.rag_initialized = True
    
    def parse_medical_text(self, text: str) -> Dict[str, str]:
        """Parse medical text"""
        medical_info = {}
        patterns = {
            'symptoms': r'Symptoms:\s*([^|]+)',
            'indications': r'Indications:\s*([^|]+)',
            'drug': r'Drug:\s*([^|]+)'
        }
        
        for field, pattern in patterns.items():
            match = re.search(pattern, text)
            medical_info[field] = match.group(1).strip() if match else ""
        
        return medical_info
    
    def predict_with_rag_and_reasoning(self, texts: List[str]) -> Dict:
        """Predict with confidence-based triple fusion"""
        if isinstance(texts, str):
            texts = [texts]
        
        if not self.rag_initialized:
            raise ValueError("RAG knowledge base not initialized")
        
        # 1. BERT predictions
        bert_results = self.predict_with_confidence(texts)
        
        # 2. RAG retrieval
        rag_results = []
        for text in texts:
            retrieval_context = self.rag_module.intelligent_retrieval(text)
            rag_results.append(retrieval_context)
        
        # 3. LangChain reasoning with RAG context
        reasoning_results = []
        for i, text in enumerate(texts):
            medical_info = self.parse_medical_text(text)
            bert_pred = bert_results['preds'][i]
            bert_conf = bert_results['confidences'][i]
            
            case_data = {
                'symptoms': medical_info['symptoms'],
                'indications': medical_info['indications'],
                'drug': medical_info['drug'],
                'bert_prediction': "Discontinue" if bert_pred == 1 else "Continue",
                'bert_confidence': bert_conf,
                'patient_factors': 'Standard patient'
            }
            
            reasoning_result = self.reasoning_agent.adaptive_reasoning(case_data)
            reasoning_results.append(reasoning_result)
        
        # 4. Confidence-based triple fusion
        enhanced_predictions = []
        enhanced_probabilities = []
        
        for i in range(len(texts)):
            # Get confidence scores
            bert_conf = bert_results['confidences'][i]
            rag_conf = rag_results[i]['retrieval_confidence']
            reasoning_conf = reasoning_results[i]['confidence']
            
            # Normalize weights by confidence
            total_conf = bert_conf + rag_conf + reasoning_conf
            weights = {
                'bert': bert_conf / total_conf,
                'rag': rag_conf / total_conf,
                'langchain': reasoning_conf / total_conf
            }
            
            # Get probability arrays
            bert_prob = bert_results['probs'][i]
            bert_prob_array = np.array([1 - bert_prob, bert_prob])
            
            rag_score = rag_results[i]['recommendation_score']
            rag_prob_array = np.array([1 - rag_score, rag_score])
            
            reasoning_decision = reasoning_results[i]['decision']
            reasoning_confidence = reasoning_results[i]['confidence']
            if reasoning_decision == 1:
                reasoning_prob_array = np.array([1 - reasoning_confidence, reasoning_confidence])
            else:
                reasoning_prob_array = np.array([reasoning_confidence, 1 - reasoning_confidence])
            
            # Weighted combination
            combined_prob = (
                weights['bert'] * bert_prob_array +
                weights['rag'] * rag_prob_array +
                weights['langchain'] * reasoning_prob_array
            )
            
            enhanced_probabilities.append(combined_prob)
            enhanced_predictions.append(np.argmax(combined_prob))
        
        return {
            'predictions': np.array(enhanced_predictions),
            'probabilities': np.array(enhanced_probabilities),
            'preds': enhanced_predictions,
            'probs': [prob[1] for prob in enhanced_probabilities],
            'bert_results': bert_results,
            'rag_results': rag_results,
            'reasoning_results': reasoning_results,
            'dynamic_weights': [{'bert': bert_results['confidences'][i], 
                               'rag': rag_results[i]['retrieval_confidence'],
                               'reasoning': reasoning_results[i]['confidence']} 
                              for i in range(len(texts))]
        }

# =============================================================================
# Multi-Chain Prompting (MCP) - Kept for SuperModel compatibility
# =============================================================================

class MultiChainPrompting:
    """Multi-chain prompting (MCP) coordinator"""
    
    def __init__(self, llm_model: str = "gpt-3.5-turbo"):
        self.llm = ChatOpenAI(model=llm_model, temperature=0.1)
        self.chains = {}
        self._setup_chains()
    
    def _setup_chains(self):
        """Setup specialized prompting chains"""
        
        # Chain A: Risk Assessment
        self.chains['risk_assessment'] = PromptTemplate(
            input_variables=['symptoms', 'drug', 'bert_prediction'],
            template="""Medical Risk Assessment Chain

Symptoms: {symptoms}
Drug: {drug}
Initial AI Prediction: {bert_prediction}

As a pharmacovigilance expert, assess the risk level:

1. Symptom severity classification (1-5 scale)
2. Drug-symptom causality probability (0-1)
3. Immediate risk level (Low/Medium/High/Critical)

Output format:
Risk Level: [Level]
Causality Score: [0.0-1.0]
Urgency: [Low/Medium/High/Critical]
Reasoning: [Brief explanation]
"""
        )
        
        # Chain B: Clinical Decision
        self.chains['clinical_decision'] = PromptTemplate(
            input_variables=['case_summary', 'similar_cases', 'risk_assessment'],
            template="""Clinical Decision Support Chain

Current Case: {case_summary}
Similar Historical Cases: {similar_cases}
Risk Assessment: {risk_assessment}

As a clinical decision specialist, provide:

1. Treatment continuation recommendation
2. Alternative options if discontinuation needed
3. Monitoring requirements
4. Patient safety considerations

Output format:
Decision: [Continue/Discontinue/Modify]
Confidence: [0.0-1.0]
Alternatives: [List if applicable]
Monitoring: [Required monitoring]
Reasoning: [Clinical justification]
"""
        )
        
        # Chain C: Evidence Integration
        self.chains['evidence_integration'] = PromptTemplate(
            input_variables=['bert_analysis', 'rag_findings', 'risk_assessment', 'clinical_decision'],
            template="""Evidence Integration and Final Decision Chain

BERT Analysis: {bert_analysis}
RAG Findings: {rag_findings}
Risk Assessment: {risk_assessment}
Clinical Decision: {clinical_decision}

As the chief medical officer, integrate all evidence and provide the final recommendation:

1. Weight each source of evidence
2. Resolve any conflicts between different analyses
3. Provide final decision with confidence level
4. Justify the decision based on medical best practices

Output format:
Final Decision: [Continue/Discontinue]
Confidence: [0.0-1.0]
Evidence Weights: [BERT: X%, RAG: Y%, Risk: Z%, Clinical: W%]
Justification: [Comprehensive medical reasoning]
Recommendations: [Follow-up actions]
"""
        )
    
    def execute_chain(self, chain_name: str, inputs: Dict) -> str:
        """Execute specified prompting chain"""
        if chain_name not in self.chains:
            raise ValueError(f"Chain '{chain_name}' not found")
        
        chain = self.chains[chain_name] | self.llm
        result = chain.invoke(inputs)
        return result.content
    
    def coordinate_chains(self, case_data: Dict) -> Dict:
        """Coordinate execution of multiple chains"""
        results = {}
        
        # Chain A: Risk Assessment
        risk_inputs = {
            'symptoms': case_data.get('symptoms', ''),
            'drug': case_data.get('drug', ''),
            'bert_prediction': case_data.get('bert_prediction', '')
        }
        results['risk_assessment'] = self.execute_chain('risk_assessment', risk_inputs)
        
        # Chain B: Clinical Decision
        clinical_inputs = {
            'case_summary': case_data.get('case_summary', ''),
            'similar_cases': case_data.get('similar_cases', ''),
            'risk_assessment': results['risk_assessment']
        }
        results['clinical_decision'] = self.execute_chain('clinical_decision', clinical_inputs)
        
        # Chain C: Evidence Integration
        integration_inputs = {
            'bert_analysis': case_data.get('bert_analysis', ''),
            'rag_findings': case_data.get('rag_findings', ''),
            'risk_assessment': results['risk_assessment'],
            'clinical_decision': results['clinical_decision']
        }
        results['evidence_integration'] = self.execute_chain('evidence_integration', integration_inputs)
        
        return results

# =============================================================================
# Model 6: Enhanced SuperModel with Dynamic Fusion
# =============================================================================

class EnhancedSuperModel(EnhancedBERTBaseline):
    """Enhanced SuperModel with dynamic fusion and all components"""
    
    def __init__(self, 
                 bert_model: str = 'bert-base-uncased',
                 embedding_model: str = "sentence-transformers/all-MiniLM-L6-v2",
                 llm_model: str = "gpt-3.5-turbo",
                 use_mcp: bool = True,
                 enable_failure_analysis: bool = True,
                 use_dynamic_fusion: bool = True):
        
        # Initialize base BERT
        super().__init__(bert_model, num_labels=2)
        
        # Initialize components
        self.rag_module = EnhancedRAGModule(embedding_model)
        self.reasoning_agent = EnhancedMedicalReasoningAgent(llm_model)
        self.mcp_coordinator = MultiChainPrompting(llm_model) if use_mcp else None
        
        # Dynamic fusion network
        self.use_dynamic_fusion = use_dynamic_fusion
        if use_dynamic_fusion:
            self.fusion_network = DynamicFusionNetwork(num_components=4).to(self.device)
            self.fusion_optimizer = torch.optim.Adam(self.fusion_network.parameters(), lr=1e-3)
        else:
            self.fusion_network = None
        
        # Configuration
        self.use_mcp = use_mcp
        self.enable_failure_analysis = enable_failure_analysis
        
        # Tracking
        self.component_performance = defaultdict(list)
        self.failure_cases = []
        self.fusion_training_data = []
        self.prediction_logs = []
        
        # State flags
        self.rag_initialized = False
        self.model_trained = False
        self.fusion_trained = False
        
        print(f"üöÄ Enhanced SuperModel initialized")
        print(f"   Dynamic Fusion: {'‚úÖ Enabled' if use_dynamic_fusion else '‚ùå Disabled'}")
        print(f"   MCP: {'‚úÖ Enabled' if use_mcp else '‚ùå Disabled'}")
        print(f"   Failure Analysis: {'‚úÖ Enabled' if enable_failure_analysis else '‚ùå Disabled'}")
    
    def setup_complete_pipeline(self, train_df: pd.DataFrame, val_df: pd.DataFrame,
                              knowledge_df: pd.DataFrame, epochs: int = 5):
        """Setup and train all components"""
        print("\n" + "="*70)
        print("üîß Setting up complete pipeline...")
        print("="*70)
        
        # 1. Train BERT model
        print("\nüìö Phase 1: Training BERT model...")
        self.train(train_df, val_df, epochs=epochs)
        self.model_trained = True
        
        # 2. Setup RAG knowledge base
        print("\nüìö Phase 2: Building RAG knowledge base...")
        self.setup_rag_knowledge_base(knowledge_df)
        
        print("\n‚úÖ Complete pipeline setup finished!")
    
    def setup_rag_knowledge_base(self, knowledge_df: pd.DataFrame):
        """Setup RAG knowledge base"""
        self.rag_module.build_knowledge_base(knowledge_df)
        self.rag_initialized = True
    
    def parse_medical_case(self, text: str) -> Dict[str, str]:
        """Parse medical case text"""
        patterns = {
            'symptoms': r'Symptoms:\s*([^|]+)',
            'indications': r'Indications:\s*([^|]+)',
            'drug': r'Drug:\s*([^|]+)',
            'age': r'Age:\s*([^|]+)',
            'sex': r'Sex:\s*([^|]+)'
        }
        
        results = {}
        for field, pattern in patterns.items():
            match = re.search(pattern, text)
            results[field] = match.group(1).strip() if match else ""
        
        results['full_text'] = text
        return results
    
    def predict_comprehensive(self, texts: Union[str, List[str]], 
                            use_all_components: bool = True,
                            return_details: bool = True) -> Dict:
        """Comprehensive prediction using all components with confidence fusion"""
        
        if not self.model_trained:
            raise ValueError("Model not trained. Call setup_complete_pipeline first.")
        
        if isinstance(texts, str):
            texts = [texts]
        
        # Check component availability
        if use_all_components and not self.rag_initialized:
            print("‚ö†Ô∏è  RAG not initialized. Using available components only.")
            use_all_components = False
        
        # Get predictions from each component
        results = self._get_comprehensive_predictions(texts, use_all_components)
        
        if return_details:
            results['final_explanations'] = []
            for i, text in enumerate(texts):
                explanation = self._generate_explanation(text, results, i)
                results['final_explanations'].append(explanation)
        
        return results
    
    def _get_comprehensive_predictions(self, texts: List[str], use_all_components: bool) -> Dict:
        """Get predictions from all available components with confidence-based fusion"""
        
        # 1. BERT predictions
        bert_results = self.predict_with_confidence(texts)
        
        # 2. RAG predictions (if available)
        rag_results = []
        if use_all_components and self.rag_initialized:
            for text in texts:
                try:
                    rag_context = self.rag_module.intelligent_retrieval(text)
                    rag_results.append(rag_context)
                except Exception as e:
                    print(f"‚ö†Ô∏è RAG error: {e}")
                    rag_results.append({'recommendation_score': 0.5, 'retrieval_confidence': 0.5})
        else:
            rag_results = [{'recommendation_score': 0.5, 'retrieval_confidence': 0.5} for _ in texts]
        
        # 3. LangChain reasoning predictions
        reasoning_results = []
        for i, text in enumerate(texts):
            try:
                medical_info = self.parse_medical_case(text)
                case_data = {
                    'symptoms': medical_info['symptoms'],
                    'indications': medical_info['indications'],
                    'drug': medical_info['drug'],
                    'bert_prediction': 'Discontinue' if bert_results['preds'][i] == 1 else 'Continue',
                    'bert_confidence': bert_results['confidences'][i],
                    'patient_factors': f"Age: {medical_info['age']}, Sex: {medical_info['sex']}"
                }
                
                reasoning_result = self.reasoning_agent.adaptive_reasoning(case_data)
                reasoning_results.append(reasoning_result)
            except Exception as e:
                print(f"‚ö†Ô∏è Reasoning error: {e}")
                reasoning_results.append({'decision': 0, 'confidence': 0.5})
        
        # 4. Dynamic confidence-based fusion
        final_predictions = []
        final_probabilities = []
        
        for i in range(len(texts)):
            # Get individual component results
            bert_prob = bert_results['probs'][i]
            bert_conf = bert_results['confidences'][i]
            
            rag_score = rag_results[i]['recommendation_score']
            rag_conf = rag_results[i]['retrieval_confidence']
            
            reasoning_decision = reasoning_results[i]['decision']
            reasoning_conf = reasoning_results[i]['confidence']
            
            # Normalize confidences for weighting
            total_conf = bert_conf + rag_conf + reasoning_conf
            
            if total_conf > 0:
                bert_weight = bert_conf / total_conf
                rag_weight = rag_conf / total_conf
                reasoning_weight = reasoning_conf / total_conf
            else:
                bert_weight = rag_weight = reasoning_weight = 1/3
            
            # Convert to probability arrays
            bert_prob_array = np.array([1 - bert_prob, bert_prob])
            rag_prob_array = np.array([1 - rag_score, rag_score])
            
            if reasoning_decision == 1:
                reasoning_prob_array = np.array([1 - reasoning_conf, reasoning_conf])
            else:
                reasoning_prob_array = np.array([reasoning_conf, 1 - reasoning_conf])
            
            # Weighted combination
            combined_prob = (
                bert_weight * bert_prob_array +
                rag_weight * rag_prob_array +
                reasoning_weight * reasoning_prob_array
            )
            
            # Add small noise to break ties
            combined_prob += np.random.normal(0, 0.01, size=2)
            combined_prob = np.clip(combined_prob, 0, 1)
            combined_prob = combined_prob / combined_prob.sum()
            
            final_probabilities.append(combined_prob)
            final_predictions.append(np.argmax(combined_prob))
        
        return {
            'predictions': np.array(final_predictions),
            'probabilities': np.array(final_probabilities),
            'preds': final_predictions,
            'probs': [prob[1] for prob in final_probabilities],
            'bert_results': bert_results,
            'rag_results': rag_results,
            'reasoning_results': reasoning_results,
            'fusion_weights': [{'bert': bert_results['confidences'][i] / (bert_results['confidences'][i] + 
                                                                          rag_results[i]['retrieval_confidence'] + 
                                                                          reasoning_results[i]['confidence']),
                              'rag': rag_results[i]['retrieval_confidence'] / (bert_results['confidences'][i] + 
                                                                               rag_results[i]['retrieval_confidence'] + 
                                                                               reasoning_results[i]['confidence']),
                              'reasoning': reasoning_results[i]['confidence'] / (bert_results['confidences'][i] + 
                                                                                rag_results[i]['retrieval_confidence'] + 
                                                                                reasoning_results[i]['confidence'])}
                             for i in range(len(texts))]
        }
    
    def _generate_explanation(self, text: str, results: Dict, index: int) -> Dict:
        """Generate explanation for a single prediction"""
        medical_case = self.parse_medical_case(text)
        
        explanation = {
            'case_summary': medical_case,
            'final_decision': results['predictions'][index],
            'final_probability': results['probabilities'][index].tolist(),
            'confidence': max(results['probabilities'][index]),
            'decision_text': 'Discontinue' if results['predictions'][index] == 1 else 'Continue',
            'component_results': {
                'bert': {
                    'prediction': results['bert_results']['preds'][index],
                    'probability': results['bert_results']['probs'][index],
                    'confidence': results['bert_results']['confidences'][index]
                },
                'rag': {
                    'recommendation_score': results['rag_results'][index]['recommendation_score'],
                    'confidence': results['rag_results'][index].get('retrieval_confidence', 0.5)
                },
                'reasoning': {
                    'decision': results['reasoning_results'][index]['decision'],
                    'confidence': results['reasoning_results'][index]['confidence']
                }
            }
        }
        
        return explanation

# =============================================================================
# Fixed Comprehensive Evaluator
# =============================================================================

class FixedComprehensiveEvaluator:
    """Fixed comprehensive model evaluator with proper interface compatibility"""
    
    def __init__(self):
        self.results = {}

    def evaluate_model(self, model, test_df: pd.DataFrame, model_name: str, sample_size: Optional[int] = None):
        """Evaluate single model with fixed interface compatibility"""
        print(f"  Evaluating {model_name}...")
    
        # Optional sampling
        if sample_size and len(test_df) > sample_size:
            test_sample = test_df.sample(n=sample_size, random_state=42)
        else:
            test_sample = test_df
    
        texts = test_sample["text"].tolist()
        labels = test_sample["label"].tolist()
    
        try:
            # Smart adaptation for different model prediction methods
            predictions, probabilities = self._get_predictions_and_probs(model, texts)
            
            # Calculate evaluation metrics
            metrics = self._calculate_metrics(labels, predictions, probabilities)
            
            # Print results
            print(f"    Accuracy:  {metrics['accuracy']:.4f}")
            print(f"    F1 Score:  {metrics['f1']:.4f}")
            print(f"    Precision: {metrics['precision']:.4f}")
            print(f"    Recall:    {metrics['recall']:.4f}")
            print(f"    ROC AUC:   {metrics['roc_auc']:.4f}")
            
            # Print confusion matrix info
            print(f"    True Pos:  {metrics.get('true_positives', 'N/A')}")
            print(f"    True Neg:  {metrics.get('true_negatives', 'N/A')}")
            print(f"    False Pos: {metrics.get('false_positives', 'N/A')}")
            print(f"    False Neg: {metrics.get('false_negatives', 'N/A')}")
    
            self.results[model_name] = {
                "metrics": metrics
            }
            
        except Exception as e:
            print(f"‚ùå Error evaluating {model_name}: {e}")
            self.results[model_name] = {"error": str(e)}
    
    def _get_predictions_and_probs(self, model, texts: List[str]) -> Tuple[List[int], List[float]]:
        """Smart adaptation for different model prediction output formats"""
        
        # Try different prediction methods in order of preference
        prediction_methods = [
            'predict_comprehensive',
            'predict_with_rag_and_reasoning',
            'predict_with_rag',
            'predict_with_reasoning',
            'predict_with_confidence'
        ]
        
        for method_name in prediction_methods:
            if hasattr(model, method_name):
                try:
                    method = getattr(model, method_name)
                    results = method(texts)
                    
                    # Handle different return formats
                    if isinstance(results, dict):
                        # Extract predictions and probabilities
                        predictions = results.get('preds', results.get('predictions', []))
                        probabilities = results.get('probs', results.get('probabilities', []))
                        
                        # Handle probability format conversion
                        if len(probabilities) > 0 and isinstance(probabilities[0], (list, np.ndarray)):
                            # 2D array: convert to class 1 probabilities
                            probs_class1 = [prob[1] if len(prob) > 1 else prob[0] for prob in probabilities]
                        else:
                            # 1D array: already class 1 probabilities
                            probs_class1 = probabilities
                            
                        return list(predictions), list(probs_class1)
                    else:
                        print(f"‚ö†Ô∏è Unexpected result format from {method_name}: {type(results)}")
                        continue
                
                except Exception as e:
                    print(f"‚ö†Ô∏è Method {method_name} failed: {e}")
                    continue
        
        # If all methods fail, raise error
        raise ValueError(f"No compatible prediction method found for model {type(model).__name__}")
    
    def _calculate_metrics(self, true_labels: List[int], predictions: List[int], 
                          probabilities: List[float]) -> Dict[str, float]:
        """Calculate evaluation metrics with confusion matrix"""
        
        # Ensure correct format
        true_labels = np.array(true_labels)
        predictions = np.array(predictions)
        probabilities = np.array(probabilities)
        
        # Calculate confusion matrix
        cm = confusion_matrix(true_labels, predictions)
        
        # Calculate basic metrics
        metrics = {
            'accuracy': accuracy_score(true_labels, predictions),
            'f1': f1_score(true_labels, predictions, zero_division=0),
            'precision': precision_score(true_labels, predictions, zero_division=0),
            'recall': recall_score(true_labels, predictions, zero_division=0),
            'true_negatives': int(cm[0, 0]) if cm.shape == (2, 2) else 0,
            'false_positives': int(cm[0, 1]) if cm.shape == (2, 2) else 0,
            'false_negatives': int(cm[1, 0]) if cm.shape == (2, 2) else 0,
            'true_positives': int(cm[1, 1]) if cm.shape == (2, 2) else 0
        }
        
        # ROC-AUC
        try:
            if len(probabilities) > 0 and len(np.unique(true_labels)) > 1:
                metrics['roc_auc'] = roc_auc_score(true_labels, probabilities)
            else:
                metrics['roc_auc'] = 0.5
        except Exception as e:
            print(f"‚ö†Ô∏è ROC-AUC calculation failed: {e}")
            metrics['roc_auc'] = 0.5
        
        return metrics

# =============================================================================
# Fixed Comprehensive Model Comparison
# =============================================================================

class FixedComprehensiveModelComparison:
    """Fixed comprehensive model comparison framework"""
    
    def __init__(self, use_large_dataset: bool = True):
        self.use_large_dataset = use_large_dataset
        self.models = {}
        self.results = {}
        self.comparison_df = None
        self.evaluator = FixedComprehensiveEvaluator()
    
    def prepare_data(self, max_samples: Optional[int] = None):
        """Prepare data with optimal size and class balancing"""
        if max_samples is None and self.use_large_dataset:
            if torch.cuda.is_available():
                available_memory = torch.cuda.get_device_properties(0).total_memory / 1024**3
                # Increase samples based on your GPU memory (16GB)
                max_samples = min(50000, int(available_memory * 3000))
            else:
                max_samples = 20000
        elif max_samples is None:
            max_samples = 10000
        
        print(f"\nüìÇ Loading up to {max_samples} samples with class balancing...")
        
        processor = EnhancedDataProcessor()
        # First load without balancing to see distribution
        df = processor.load_and_prepare_data(sample_size=max_samples, balance_classes=False)
        
        # Check if we have enough data
        label_counts = df['label'].value_counts()
        print(f"\nüìä Original label distribution:")
        print(f"   Class 0 (No dechallenge): {label_counts.get(0, 0)} samples")
        print(f"   Class 1 (Dechallenge): {label_counts.get(1, 0)} samples")
        
        # Balance classes with different strategies based on data size
        minority_size = label_counts.min()
        
        if minority_size < 2000:
            print(f"\n‚ö†Ô∏è  Limited minority class samples ({minority_size}). Using enhanced balancing strategy.")
            # For severe imbalance, use more aggressive balancing
            # Allow up to 3:1 ratio instead of 2:1
            balanced_dfs = []
            for label in [0, 1]:
                label_df = df[df["label"] == label]
                if len(label_df) > minority_size * 3:
                    label_df = label_df.sample(n=minority_size * 3, random_state=42)
                balanced_dfs.append(label_df)
            
            df = pd.concat(balanced_dfs, ignore_index=True)
            df = df.sample(frac=1, random_state=42).reset_index(drop=True)
            
            print(f"‚úÖ Applied 3:1 class balancing")
            print(f"   New distribution: {df['label'].value_counts().to_dict()}")
        else:
            # Standard balancing
            df = processor.load_and_prepare_data(sample_size=max_samples, balance_classes=True)
        
        train_df, val_df, test_df = processor.create_enhanced_splits(df)
        
        # For knowledge base, use all available training data
        knowledge_df = pd.concat([train_df, val_df]).sample(
            n=min(10000, len(train_df) + len(val_df)), 
            random_state=42
        )
        
        return train_df, val_df, test_df, knowledge_df
    
    def initialize_all_models(self):
        """Initialize all model variants"""
        print("\nüèóÔ∏è Initializing all models...")
        
        # 1. BERT Baseline
        self.models['BERT_Baseline'] = EnhancedBERTBaseline()
        
        # 2. Biomedical BERTs - restore all variants
        biomedical_models = ['BioBERT', 'ClinicalBERT', 'SciBERT']
        for model_type in biomedical_models:
            try:
                self.models[model_type] = BiomedicalBERT(model_type)
                print(f"‚úÖ {model_type} initialized")
            except Exception as e:
                print(f"‚ùå Failed to initialize {model_type}: {e}")
        
        # 3. BERT + RAG
        self.models['BERT_RAG'] = BERTWithRAG()
        
        # 4. Models requiring API key
        if os.getenv("OPENAI_API_KEY"):
            self.models['BERT_LangChain'] = BERTWithLangChain()
            self.models['BERT_RAG_LangChain'] = BERTWithRAGAndLangChain()
            self.models['SuperModel_Static'] = EnhancedSuperModel(use_dynamic_fusion=False)
            self.models['SuperModel_Dynamic'] = EnhancedSuperModel(use_dynamic_fusion=True)
            print("‚úÖ LangChain-based models initialized")
        else:
            print("‚ö†Ô∏è  No OpenAI API key - LangChain models skipped")
    
    def train_all_models(self, train_df, val_df, knowledge_df, epochs=5):
        """Train all initialized models"""
        print("\nüîß Training all models...")
        
        for name, model in self.models.items():
            print(f"\nüìä Training {name}...")
            
            try:
                # Train base model
                model.train(train_df, val_df, epochs=epochs)
                
                # Setup RAG if applicable
                if hasattr(model, 'build_rag_knowledge_base'):
                    model.build_rag_knowledge_base(knowledge_df)
                elif hasattr(model, 'setup_complete_pipeline'):
                    # For SuperModel
                    model.setup_complete_pipeline(train_df, val_df, knowledge_df, epochs=epochs)
                
                print(f"‚úÖ {name} trained successfully")
                
            except Exception as e:
                print(f"‚ùå Error training {name}: {e}")
                self.models[name] = None
    
    def evaluate_all_models(self, test_df, sample_size: Optional[int] = None):
        """Evaluate all models on test set"""
        print("\nüìä Evaluating all models...")
        
        # Sample test set if needed
        if sample_size and len(test_df) > sample_size:
            test_sample = test_df.sample(n=sample_size, random_state=42)
        else:
            test_sample = test_df
        
        print(f"  Using {len(test_sample)} test samples...")
        print(f"  Test set composition: {test_sample['label'].value_counts().to_dict()}")
        
        for name, model in self.models.items():
            if model is None:
                print(f"‚è≠Ô∏è Skipping {name} (not initialized)")
                continue
                
            try:
                self.evaluator.evaluate_model(model, test_sample, name)
                
            except Exception as e:
                print(f"‚ùå Error evaluating {name}: {e}")
                self.evaluator.results[name] = {'error': str(e)}
        
        # Update results and create comparison DataFrame
        self.results = self.evaluator.results
        self._create_comparison_dataframe()
    
    def _create_comparison_dataframe(self):
        """Create comparison DataFrame from results"""
        valid_results = {}
        
        for name, result in self.results.items():
            if isinstance(result, dict) and 'metrics' in result:
                valid_results[name] = result['metrics']
        
        if valid_results:
            self.comparison_df = pd.DataFrame(valid_results).T
            
            # Add model categories
            categories = []
            for name in self.comparison_df.index:
                if 'SuperModel' in name:
                    categories.append('SuperModel')
                elif any(bio in name for bio in ['Bio', 'Clinical', 'Sci', 'PubMed']):
                    categories.append('Biomedical')
                elif 'RAG' in name and 'LangChain' in name:
                    categories.append('Hybrid')
                elif 'RAG' in name:
                    categories.append('RAG-Enhanced')
                elif 'LangChain' in name:
                    categories.append('LLM-Enhanced')
                else:
                    categories.append('Baseline')
            
            self.comparison_df['category'] = categories
    
    def generate_report(self):
        """Generate comprehensive comparison report"""
        print("\n" + "="*70)
        print("üìã COMPREHENSIVE MODEL COMPARISON REPORT")
        print("="*70)
        
        if self.comparison_df is not None and len(self.comparison_df) > 0:
            # Overall comparison
            print("\nüìä Performance Metrics:")
            print(self.comparison_df[['accuracy', 'f1', 'precision', 'recall', 'roc_auc']].round(4))
            
            # Best models by metric
            print("\nüèÜ Best Models by Metric:")
            for metric in ['accuracy', 'f1', 'precision', 'recall', 'roc_auc']:
                if metric in self.comparison_df.columns:
                    best_model = self.comparison_df[metric].idxmax()
                    best_score = self.comparison_df.loc[best_model, metric]
                    print(f"  {metric.upper()}: {best_model} ({best_score:.4f})")
            
            # Category comparison
            print("\nüìä Average Performance by Category:")
            category_avg = self.comparison_df.groupby('category')[['accuracy', 'f1', 'precision', 'recall', 'roc_auc']].mean()
            print(category_avg.round(4))
            
            # Improvement analysis
            if 'BERT_Baseline' in self.comparison_df.index:
                print("\nüìà Improvement over BERT Baseline:")
                baseline_metrics = self.comparison_df.loc['BERT_Baseline']
                
                for model in self.comparison_df.index:
                    if model != 'BERT_Baseline':
                        print(f"\n  {model}:")
                        for metric in ['accuracy', 'f1', 'roc_auc']:
                            if metric in self.comparison_df.columns:
                                improvement = ((self.comparison_df.loc[model, metric] - 
                                              baseline_metrics[metric]) / 
                                             baseline_metrics[metric] * 100)
                                print(f"    {metric}: {improvement:+.2f}%")
            
            # Create visualizations
            self._create_visualizations()
        
        # Failed models
        failed_models = [name for name, result in self.results.items() 
                        if isinstance(result, dict) and 'error' in result]
        if failed_models:
            print("\n‚ö†Ô∏è  Failed Models:")
            for model in failed_models:
                print(f"  {model}: {self.results[model]['error']}")
        
        # GPU usage
        if torch.cuda.is_available():
            print(f"\nüñ•Ô∏è  GPU Memory Usage:")
            print(f"  Allocated: {torch.cuda.memory_allocated(0) / 1024**3:.2f} GB")
            print(f"  Reserved: {torch.cuda.memory_reserved(0) / 1024**3:.2f} GB")
    
    def _create_visualizations(self):
        """Create comparison visualizations"""
        if self.comparison_df is None or len(self.comparison_df) == 0:
            return
        
        # Create figure with subplots
        fig, axes = plt.subplots(2, 3, figsize=(18, 12))
        fig.suptitle('Comprehensive Model Comparison', fontsize=16)
        
        # 1. Accuracy comparison
        ax = axes[0, 0]
        self.comparison_df['accuracy'].plot(kind='bar', ax=ax, color='skyblue')
        ax.set_title('Accuracy Scores')
        ax.set_ylabel('Accuracy')
        ax.set_ylim(0, 1.1)
        ax.tick_params(axis='x', rotation=45)
        
        # 2. F1 Score comparison
        ax = axes[0, 1]
        self.comparison_df['f1'].plot(kind='bar', ax=ax, color='lightgreen')
        ax.set_title('F1 Scores')
        ax.set_ylabel('F1 Score')
        ax.set_ylim(0, 1.1)
        ax.tick_params(axis='x', rotation=45)
        
        # 3. ROC-AUC comparison
        ax = axes[0, 2]
        self.comparison_df['roc_auc'].plot(kind='bar', ax=ax, color='salmon')
        ax.set_title('ROC-AUC Scores')
        ax.set_ylabel('ROC-AUC')
        ax.set_ylim(0, 1.1)
        ax.tick_params(axis='x', rotation=45)
        
        # 4. Precision-Recall scatter
        ax = axes[1, 0]
        ax.scatter(self.comparison_df['recall'], self.comparison_df['precision'], s=100)
        for idx, model in enumerate(self.comparison_df.index):
            ax.annotate(model, (self.comparison_df.loc[model, 'recall'], 
                              self.comparison_df.loc[model, 'precision']),
                       fontsize=8, rotation=15)
        ax.set_xlabel('Recall')
        ax.set_ylabel('Precision')
        ax.set_title('Precision vs Recall')
        ax.grid(True, alpha=0.3)
        
        # 5. Category performance
        ax = axes[1, 1]
        category_avg = self.comparison_df.groupby('category')['f1'].mean()
        category_avg.plot(kind='bar', ax=ax, color='orchid')
        ax.set_title('Average F1 by Model Category')
        ax.set_ylabel('Average F1 Score')
        ax.tick_params(axis='x', rotation=45)
        
        # 6. Model complexity comparison
        ax = axes[1, 2]
        complexity_scores = []
        complexity_labels = []
        for model in self.comparison_df.index:
            if 'SuperModel' in model:
                complexity_scores.append(4)
            elif 'RAG' in model and 'LangChain' in model:
                complexity_scores.append(3)
            elif 'RAG' in model or 'LangChain' in model:
                complexity_scores.append(2)
            else:
                complexity_scores.append(1)
            complexity_labels.append(model)
        
        colors = ['blue', 'green', 'orange', 'red']
        ax.bar(range(len(complexity_labels)), complexity_scores, 
               color=[colors[score-1] for score in complexity_scores])
        ax.set_title('Model Complexity')
        ax.set_ylabel('Complexity Level')
        ax.set_xticks(range(len(complexity_labels)))
        ax.set_xticklabels(complexity_labels, rotation=45)
        
        plt.tight_layout()
        plt.savefig('comprehensive_model_comparison.png', dpi=300, bbox_inches='tight')
        print("\nüìä Visualizations saved to 'comprehensive_model_comparison.png'")

# =============================================================================
# Main Execution Function
# =============================================================================

def run_fixed_comparison(max_samples: int = 50000):
    """Run the complete fixed model comparison"""
    
    print("üîß Starting Fixed Model Comparison...")
    print("="*70)
    
    try:
        # Create fixed comparator
        comparator = FixedComprehensiveModelComparison(use_large_dataset=True)
        
        # 1. Data preparation - with larger sample size
        print("üìä Phase 1: Data Preparation")
        train_df, val_df, test_df, knowledge_df = comparator.prepare_data(max_samples=max_samples)
        
        # 2. Model initialization
        print("\nüèóÔ∏è Phase 2: Model Initialization")
        comparator.initialize_all_models()
        
        # 3. Model training - reduced epochs for testing
        print("\nüéØ Phase 3: Model Training")
        comparator.train_all_models(train_df, val_df, knowledge_df, epochs=3)
        
        # 4. Model evaluation
        print("\nüìä Phase 4: Model Evaluation")
        comparator.evaluate_all_models(test_df, sample_size=min(2000, len(test_df)))
        
        # 5. Report generation
        print("\nüìã Phase 5: Report Generation")
        comparator.generate_report()
        
        print("\n‚úÖ Complete model comparison finished successfully!")
        return comparator
        
    except Exception as e:
        print(f"‚ùå Error in model comparison: {e}")
        import traceback
        traceback.print_exc()
        return None

# =============================================================================
# Example Usage and Testing
# =============================================================================

def test_individual_models():
    """Test individual models to ensure they work correctly"""
    
    print("üß™ Testing individual models...")
    
    # Create sample data
    sample_data = {
        'text': [
            'Symptoms: Nausea; Vomiting | Indications: Inflammation | Drug: Prednisolone | Age: 45 | Sex: M',
            'Symptoms: Headache; Dizziness | Indications: Asthma | Drug: Prednisolone | Age: 32 | Sex: F'
        ],
        'label': [1, 0]
    }
    sample_df = pd.DataFrame(sample_data)
    
    # Test BERT baseline
    print("\n1. Testing BERT Baseline...")
    try:
        bert_model = EnhancedBERTBaseline()
        predictions = bert_model.predict_with_confidence(sample_data['text'])
        print(f"   ‚úÖ BERT predictions: {predictions['preds']}")
        print(f"   ‚úÖ BERT probabilities: {predictions['probs']}")
    except Exception as e:
        print(f"   ‚ùå BERT test failed: {e}")
    
    # Test evaluator
    print("\n2. Testing Evaluator...")
    try:
        evaluator = FixedComprehensiveEvaluator()
        evaluator.evaluate_model(bert_model, sample_df, "Test_BERT")
        print(f"   ‚úÖ Evaluator results: {evaluator.results}")
    except Exception as e:
        print(f"   ‚ùå Evaluator test failed: {e}")
    
    print("\n‚úÖ Individual model tests completed!")

if __name__ == "__main__":
    # Run the complete fixed comparison
    print("üöÄ Starting Complete FAERS Prednisolone Model Comparison")
    print("="*70)
    
    # Check data availability first
    import os
    if not os.path.exists("./final_processed_data.csv"):
        print("‚ùå Error: final_processed_data.csv not found!")
        print("   Please ensure your data file is in the current directory.")
    else:
        # Check data size
        temp_df = pd.read_csv("./final_processed_data.csv")
        total_records = len(temp_df[temp_df["dechal"].isin(["Y", "N"])])
        print(f"üìä Total valid records available: {total_records}")
        
        if total_records < 10000:
            print(f"‚ö†Ô∏è  Warning: Limited data ({total_records} records). Model performance may be affected.")
            print("   Consider using data augmentation or collecting more samples.")
        
        # Run comparison with appropriate sample size
        max_samples = min(50000, total_records)
        comparator = run_fixed_comparison(max_samples=max_samples)
        
        if comparator:
            print("\nüéâ Comparison completed successfully!")
            if comparator.comparison_df is not None:
                print("\nüìä Final Results Summary:")
                print(comparator.comparison_df[['accuracy', 'f1', 'precision', 'recall', 'roc_auc']].round(4))
        else:
            print("\n‚ùå Comparison failed. Please check the error messages above.")
```
